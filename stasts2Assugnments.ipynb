Median: The median is the middle value of a dataset when the values are arranged in ascending or descending order. If there is an even number of values, the median is the average of the two middle values.

Mode: The mode is the value that appears most frequently in a dataset. It is possible for a dataset to have one mode (unimodal), two modes (bimodal), or more than two modes (multimodal), or to have no mode if all values occur with the same frequency.

Mean: The mean is the arithmetic average of a dataset. It is calculated by adding up all the values in the dataset and then dividing by the total number of values. The mean is sensitive to extreme values (outliers) in the dataset because it takes into account the magnitude of each value.

If there is an even number of values, the median is the average of the two middle values. The median is not influenced by extreme values and is often used when the dataset contains outliers or when the distribution of data is skewed.

Mode: The mode is the value that appears most frequently in a dataset. Unlike the mean and median, the mode does not take into account the magnitude of the values, only their frequency of occurrence. The mode is useful for categorical or discrete data and can be used to identify the most common category or value in a dataset.

A Venn diagram is a graphical representation that illustrates the relationships and similarities/differences between different sets of data or groups. It consists of overlapping circles, each representing a set, with the overlapping regions showing the intersections between the sets. The non-overlapping regions represent the elements that are unique to each set.

Venn diagrams are widely used in various fields, including mathematics, logic, statistics, computer science, and more, to visualize relationships between different categories or groups of data. They are particularly useful for understanding concepts such as union, intersection, complement, and subset relationships between sets.

Venn diagrams are named after John Venn, a British logician and philosopher who introduced them in the late 19th century as a tool for illustrating logical relationships between different sets. They provide a clear and concise way to organize and visualize information, making complex relationships easier to understand at a glance.

Positive Skewness (Right Skewed): In a positively skewed distribution, the tail of the distribution extends towards the right, and the majority of the data points are concentrated on the left side. This means that there are more extreme values on the right side of the distribution, pulling the mean to the right of the median.

Negative Skewness (Left Skewed): In a negatively skewed distribution, the tail of the distribution extends towards the left, and the majority of the data points are concentrated on the right side. This means that there are more extreme values on the left side of the distribution, pulling the mean to the left of the median.

Zero Skewness: A distribution is considered to have zero skewness when it is perfectly symmetrical, with the mean, median, and mode all equal and located at the center of the distribution.
